```python
import pandas as pd
from os import getcwd
path=getcwd()
df_raw1 = pd.read_csv(path+'/data/raw/'+'Estacao_Convencional.csv')
df_raw2 = pd.read_csv(path+'/data/raw/'+'Sensor_FieldPRO.csv')
df_raw2.head(n=10)
```

```python
na_raw2 =list(df_raw2[df_raw2.isna().any(axis=1)].index)
print(df_raw2.iloc[:32,:3])
```

## Cleaning the NA using mean of the the before and after

```python
df_mod1 = df_raw2[:]
for row in na_raw2:
    df_mod1.iloc[row,1:3] = df_mod1.iloc[[row-1,row+1],1:3].mean()
    print(df_mod1.iloc[row,1:3])
```

```python
df_mod1.head(n=10)
```

# Joining the two tables

## Cleaning and joing tables using date

```python
import pytz
df_mod1['Datetime – utc'] = pd.to_datetime(df_mod1['Datetime – utc'])
br = pytz.timezone('America/Sao_Paulo')
df_mod1['Datetime – BR'] = pd.to_datetime(df_mod1['Datetime – utc'].dt.strftime('%Y-%m-%d %H'),utc=True)
df_mod1['Datetime – BR'] = df_mod1['Datetime – BR'].dt.tz_convert(br)
df_mod1.head()
```

```python
df_mod2 = df_raw1.copy()
```

```python
df_mod2['Datetime – BR'] = df_mod2['data'] + ' ' + df_mod2['Hora (Brasília)']
df_mod2.loc[:,'Datetime – BR'] = pd.to_datetime(df_mod2.loc[:,'Datetime – BR'],format='%Y-%m-%d %H')
df_mod2.loc[:,'Datetime – BR'] = df_mod2['Datetime – BR'].dt.tz_localize(br)
df_mod2.head()
```

```python
df1= df_mod1.merge(right=df_mod2,how="left",on='Datetime – BR')
df1.head()
```

```python
print(df1.isnull().sum())
```

```python
 df1[df1.isnull().any(axis=1)]
### Nulls values that join produced
```

```python
df_1_na = df1[df1.isnull().any(axis=1)]
df_1_na['Datetime – BR'].describe(datetime_is_numeric=True)
### This data will be used for testing data.
```

```python
drop_col = ['Datetime – utc','data','Hora (Brasília)']
df_1_na = df_1_na.drop(drop_col,axis=1)
df_1_na.to_csv(path+'/data/'+'real_data_for_testing.csv')
```

```python
df =  df1[~df1.isnull().any(axis=1)]
df = df.drop(drop_col,axis=1)
df.to_csv(path+'/data/'+'data_for_modeling.csv')
###Finishing cleaning and saving 
```
